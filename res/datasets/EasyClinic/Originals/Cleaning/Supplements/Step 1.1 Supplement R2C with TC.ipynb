```python
import pandas as pd
import numpy as np
import os
import sys
```

```python
sys.path.append("../../../../Tracer/src")
from DataHandling.TraceCreator import create_trace_matrix_from_trace_list
```

```python
def create_trace_list(path_to_oracle):
    oracle = open(path_to_oracle).read().split("\n")
    trace_list = []
    for trace_item in oracle:
        if len(trace_item) > 0:
            source, targets = trace_item.split(":")
            for target in targets.split(" "):
                if len(target) != 0:
                    trace_list.append(" ".join([source, target]))
                else:
                    trace_list.append(" ".join([source, ""]))
    return create_trace_matrix_from_trace_list(trace_list)
```

# 1 -> 2

```python
path_to_oracle = "../Originals/TracedMatrices/UC_TC.txt"
uc_tc = create_trace_list(path_to_oracle)
print("shape: ", uc_tc.shape)
uc_tc.head(1)
```

# 2 -> 3

```python
path_to_oracle = "../Originals/TracedMatrices/ID_TC.txt"
id_tc = create_trace_list(path_to_oracle)
print("shape: ", id_tc.shape)
id_tc.head(1)
```

```python
trace_list = []
common_columns = set(uc_tc.columns) & set(id_tc.columns)
for col in common_columns:
    
    a = uc_tc[uc_tc[col] == 1].index
    b = id_tc[id_tc[col] == 1].index    
    
    for a_id in a:
        for b_id in b:
            trace_list.append(" ".join([a_id, b_id]))
```

```python
oracle = pd.read_csv("../../Oracles/DataFrames/Level_1_to_Level_2.csv").set_index("id")
```

```python
for trace in trace_list:
    id_id, uc_id = trace.split(" ")
    if uc_id in list(oracle.index):
        if oracle.loc[uc_id][id_id] == 0:
            print("Missed a traced :(")
```
